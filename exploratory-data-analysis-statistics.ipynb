שאלה 1: האם תוחלת רמת החומצה של שני סוגי היינות: היין מזן 1 והיין מזן 2, שוות אחת לשנייה ברמת מובהקות של 5%

H0-הפרש התוחלות בין תוחלת רמת החומצה של יין מזן 1 לתוחלת של יין מזן 2 = 0 כלומר התוחלות של רמת החומצה של שני סוגי היינות שווה 


H1-הפרש התוחלות בין תוחלת של יין מזן 1 לתוחלת של יין מזן 2 לא שווה ל-0

שאלה 2: האם יש הבדל בין תוחלת רמת האפר ביינות מזן 1 לבין תוחלת רמת האפר ביינות מזן 2 ברמת מובהקות של 
2% 

H0- הפרש התוחלות = 0

H1- הפרש התוחלות != 0

```python
mean_n1 =cultivars_1['Malic acid'].mean()
mean_n2= cultivars_2['Malic acid'].mean()

# התיחסתי לסטיית התקן כסטיית תקן מדגמית 
#Sp(((n1 -1)S1**2 + (n2-1)S2**2) / (n1 + n2 -2))**0.5

std_n1= cultivars_1['Malic acid'].std()
std_n2=  cultivars_2['Malic acid'].std()
sample_std_N = ((std_n1**2)*len(cultivars_1 -1)+ (std_n2**2)*len(cultivars_2 -1)) / (len(cultivars_2) + len(cultivars_2) - 2)**0.5
print('Sp=',sample_std_N)
T_sample= (mean_n1 - mean_n2 -0) / (sample_std_N * (1/len(cultivars_1) + 1/len(cultivars_2))**0.5)
print('T_sample=',T_sample)

alpha = 0.05  # Significance level
# 2-sample Z-test
# Z_test ---> nomral_p (n>30) 
#בהנחה שסטיית התקן המדגמית שווה לסטיית התקן של האוכלוסיה
# מחשבים את הממוצע המשוער של שתי האוכלוסיות

mean_difference = (mean_n1 -mean_n2)
print('mean difference = ',mean_difference)
stddev_difference = np.sqrt((std_n1**2/len(cultivars_1)) + (std_n2**2/len(cultivars_2)))
z_score = mean_difference / stddev_difference
p_value_z = 2 * (1 - norm.cdf(abs(z_score)))
print("P_VALUE:", p_value_z)

if p_value_z < alpha:
    print("we reject H0 according to the normal distribution " + "יש הבדל סטטיסטי בין שתי האוכלוסיות")
else:
    print("we don't reject H0 according to the normal distribution" + "אין הבדל סטטיסטי בין שתי האוכלוסיות")


#########################################################################
print("##########")
# 2-sample t-test
# T_test ---> T(sample1_mean - sample2_mean) 



t_statistic, p_value = stats.ttest_ind(cultivars_1['Malic acid'], cultivars_2['Malic acid'])
t_critical_left = stats.t.ppf(alpha / 2, df=(len(cultivars_2) + len(cultivars_2) - 2))
t_critical_right = -t_critical_left 



print("t-statistic:", t_statistic)
print("p-value:", p_value)
print("t-critical left:", t_critical_left)
print("t-critical right:", t_critical_right)


if (t_statistic < t_critical_left or t_statistic > t_critical_right) :
    print("we reject H0 according to the T distribution (T-Test) " + "יש הבדל סטטיסטי בין שתי האוכלוסיות")
else:
    print("we don't reject H0 according to the T distribution (T-Test)" +"אין הבדל סטטיסטי בין שתי האוכלוסיות")


```
